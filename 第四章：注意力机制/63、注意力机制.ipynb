# 注意力汇聚：Nadaraya-Watson 核回归


```python
# 下面两行是防止内核挂掉加的
import os
os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"

import torch
from torch import nn
from d2l import torch as d2l
```

#### 生成数据集

```python
n_train = 50
x_train, _ = torch.sort(torch.rand(n_train) * 5) # 训练数据，并排序

def f(x): # 生成真实数据
    return 2 * torch.sin(x) + x**0.8

y_train = f(x_train) + torch.normal(0.0, 0.5, (n_train,)) # 真实的label,这里加噪音了
x_test = torch.arange(0, 5, 0.1) # 测试集
y_truth = f(x_test) # 测试的真实标签
n_test = len(x_test)
n_test
```

```python
def plot_kernel_reg(y_hat):
    d2l.plot(x_test, [y_truth, y_hat], 'x', 'y', legend=['Truth', 'Pred'],
             xlim=[0, 5], ylim=[-1, 5])
    d2l.plt.plot(x_train, y_train, 'o', alpha=0.5);

y_hat = torch.repeat_interleave(y_train.mean(), n_test) # 最简单的预测就是：不管输入咋样，都取平均
plot_kernel_reg(y_hat) 
```

显然上面的程序不管x的输入，把所有的输出都用输入的平均代替，是有问题的，下面看看 非参数注意力汇聚
#### 非参数注意力汇聚

直观的理解就是：把每个输入附近的数据拿过来，得到输出

```python
X_repeat = x_test.repeat_interleave(n_train).reshape((-1, n_train)) # 测试数据
attention_weights = nn.functional.softmax(-(X_repeat - x_train)**2 / 2, dim=1) # 计算测试数据和训练数据 之间 的距离，得到注意力权重
y_hat = torch.matmul(attention_weights, y_train) # 通过训练标签，得到预测标签
plot_kernel_reg(y_hat)
```

#### 注意力权重：给定一个训练输入，看注意力关注测试输入的哪里（越红表示越关注）

```python
d2l.show_heatmaps(
    attention_weights.unsqueeze(0).unsqueeze(0),
    xlabel='Sorted training inputs', ylabel='Sorted testing inputs')
```

#### 带参数注意力汇聚 假定两个张量的形状分别是 (n,a,b) 和 (n,b,c) ，它们的批量矩阵乘法输出的形状为 (n,a,c)

```python
X = torch.ones((2, 1, 4))
Y = torch.ones((2, 4, 6))
torch.bmm(X, Y).shape # bmm就是带batch的矩阵乘法，也就是说：把b不动，只乘后面的
```

#### 使用小批量矩阵乘法来计算小批量数据中的加权平均值

```python
weights = torch.ones((2, 10)) * 0.1
values = torch.arange(20.0).reshape((2, 10))
torch.bmm(weights.unsqueeze(1), values.unsqueeze(-1))
```

#### 带参数的注意力汇聚

```python
class NWKernelRegression(nn.Module):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        self.w = nn.Parameter(torch.rand((1,), requires_grad=True))

    def forward(self, queries, keys, values):
        queries = queries.repeat_interleave(keys.shape[1]).reshape(
            (-1, keys.shape[1]))
        self.attention_weights = nn.functional.softmax(
            -((queries - keys) * self.w)**2 / 2, dim=1) # 和前面不同的地方：带了一个可学习的参数
        return torch.bmm(self.attention_weights.unsqueeze(1),
                         values.unsqueeze(-1)).reshape(-1)
```

#### 将训练数据集转换为键和值

```python
X_tile = x_train.repeat((n_train, 1))
Y_tile = y_train.repeat((n_train, 1))
keys = X_tile[(1 - torch.eye(n_train)).type(torch.bool)].reshape(
    (n_train, -1))
values = Y_tile[(1 - torch.eye(n_train)).type(torch.bool)].reshape(
    (n_train, -1))
```

#### 训练带参数的注意力汇聚模型

```python
net = NWKernelRegression()
loss = nn.MSELoss(reduction='none')
trainer = torch.optim.SGD(net.parameters(), lr=0.5)
animator = d2l.Animator(xlabel='epoch', ylabel='loss', xlim=[1, 5])

for epoch in range(5):
    trainer.zero_grad()
    l = loss(net(x_train, keys, values), y_train) / 2
    l.sum().backward()
    trainer.step()
    print(f'epoch {epoch + 1}, loss {float(l.sum()):.6f}')
    animator.add(epoch + 1, float(l.sum()))
```

#### 预测结果绘制

```python
keys = x_train.repeat((n_test, 1))
values = y_train.repeat((n_test, 1))
y_hat = net(x_test, keys, values).unsqueeze(1).detach()
plot_kernel_reg(y_hat)
```
